```python
# 2. 미분하고 그래프 그리기
```

```python
import numpy as np
import matplotlib.pylab as plt

h = 0.0000000001
def numerical_diff(f, x):
    return (f(x+h) - f(x-h)) / (2*h)

def f1(x):
    return x**2
def df1(x):
    return numerical_diff(f1, x)

def f2(x):
    return x**3 + x**2 + x 
def df2(x):
    return numerical_diff(f2, x)

def f3(x):
    return x*np.sin(x*x*x+3)
def df3(x):
    return numerical_diff(f3, x)

def f4(x):
    return np.cos(np.sqrt(np.exp(x) + 1) / 2)
def df4(x):
    return numerical_diff(f4, x)

print()
print("#1. x에 값 대입하기")
    
print("f1(2) =", f1(2))
print("df1(2) =", round(df1(2), 4))
print("f2(2) =", f2(2))
print("df2(2) =", round(df2(2), 4))
print("f3(2) =", round(f3(2), 4))
print("df3(2) =", round(df3(2), 4))
print("f4(2) =", round(f4(2), 4))
print("df4(2) =", round(df4(2), 4))

print()
print("#2. 그래프 그리기")
x = np.arange(-5.0, 5.0, 0.1)
plt.ylim(-20, 20)

y = f1(x)
dy = df1(x)
plt.plot(x, dy)
plt.plot(x, y)

y = f2(x)
dy = df2(x)
plt.plot(x, dy)
plt.plot(x, y)

y = f3(x)
dy = df3(x)
plt.plot(x, dy)
plt.plot(x, y)

y = f4(x)
dy = df4(x)
plt.plot(x, dy)
plt.plot(x, y)
```
