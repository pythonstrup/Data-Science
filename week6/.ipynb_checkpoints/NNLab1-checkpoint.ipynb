```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

i = np.array([0.9, 0.1, 0.8], float).reshape(3,1)
w_input_hidden = np.array([0.9, 0.3, 0.4, 0.2, 0.8, 0.2, 0.1, 0.5, 0.6]).reshape(3,3)
w_hidden_output = np.array([.3, .7, .5, .6, .5, .2, .8, .1, .9]).reshape(3,3)

i
w_input_hidden
w_hidden_output
```

```python
x_hidden = np.dot(w_input_hidden, i)
x_hidden

o_hidden = sigmoid(x_hidden)
o_hidden
```

```python
x_output = np.dot(w_hidden_output, o_hidden)
x_output

o_output = sigmoid(x_output)
o_output
```

```python
# 3. 222 신경망 업데이트 계산
```

```python
e_output1 = 0.8
e_output2 = 0.5

w11 = 2.0
w12 = 1.0
w21 = 3.0
w22 = 4.0

o1 = 0.4
o2 = 0.5

e = np.array([.8, .5]).reshape(2,1)
w = np.array([w11, w21, w12, w22]).reshape(2,2)
o = np.array([o1, o2]).reshape(2, 1)
s = np.dot(w, o)
```

```python
print("**행렬곱 계산**")
# 행렬곱 계산
value = -e * sigmoid(s) * (1-sigmoid(s))
w_update = np.dot(value, o.T)
w_update

# new = old - w변화량
print("실제 업데이트값")
w - w_update

print()
print("**단순연산**")
# 단순 계산
w11_update = -e_output1 * sigmoid(w11 * o1 + w21 * o2) * (1-sigmoid(w11 * o1 + w21 * o2)) * o1
w12_update = -e_output2 * sigmoid(w12 * o1 + w22 * o2) * (1-sigmoid(w12 * o1 + w22 * o2)) * o1
w21_update = -e_output1 * sigmoid(w11 * o1 + w21 * o2) * (1-sigmoid(w11 * o1 + w21 * o2)) * o2
w22_update = -e_output2 * sigmoid(w12 * o1 + w22 * o2) * (1-sigmoid(w12 * o1 + w22 * o2)) * o2

# 가중치 업데이트
# new = old - w변화량
print("은닉계층 업데이트 결과")
w11 - w11_update
w12 - w12_update
w21 - w21_update
w22 - w22_update
```

```python
# 입력계층까지 도전
w11 = 3.0
w12 = 1.0
w21 = 2.0
w22 = 7.0

# 2번 문제 은닉계층 오차값
# e = e_hidden
e= np.array([[0.42],[0.88]])

# 가중치
w = np.array([w11, w21, w12, w22]).reshape(2,2)

# 주어진 입력값
o = np.array([-0.14938188, 0.02134027]).T

# 시그모이드 대입 함수값
s = np.dot(w, o)

print("w업데이트")
value = -e * sigmoid(s) * (1-sigmoid(s))
w_update = np.dot(value, o.T)
w_update

# new = old - w변화량
print("실제 업데이트된 값")
w - w_update
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

i = np.array([0.9, 0.1, 0.8], float).reshape(3,1)
# 입력->은닉 가중치
w_input_hidden = np.array([0.9, 0.3, 0.4, 0.2, 0.8, 0.2, 0.1, 0.5, 0.6]).reshape(3,3)
# 은닉->출력 가중치
w_hidden_output = np.array([.3, .7, .5, .6, .5, .2, .8, .1, .9]).reshape(3,3)

i
w_input_hidden
w_hidden_output
```

```python
print("은닉계층 output(1번 문제 참고)")
o_hidden
print()

print("출력부 output(1번 문제 참고)")
o_output

print("목표치")
targets = np.array([.01, .01, .99]).reshape(3,-1)
targets
```

```python
print("출력 오류값")
e_output = targets - o_output
e_output
```

```python
# 에러값
e1 = e_output[0][0]
e2 = e_output[1][0]
e3 = e_output[2][0]

# 가중치
w11 = w_hidden_output[0][0]
w12 = w_hidden_output[1][0]
w13 = w_hidden_output[2][0]
w21 = w_hidden_output[0][1]
w22 = w_hidden_output[1][1]
w23 = w_hidden_output[2][1]
w31 = w_hidden_output[0][2]
w32 = w_hidden_output[1][2]
w33 = w_hidden_output[2][2]

# 은닉계층 오차 계산 - 행렬곱
print("**입력계층 오차 - 행렬곱**")
error = np.array([e1, e2, e3]).reshape(3,1)
normalization = np.array([w11 / (w11 + w21 + w31), w12 / (w12 + w22 + w32), w13 / (w13 + w23 + w33),
                          w21 / (w11 + w21 + w31), w22/ (w12 + w22 + w32), w23 / (w13 + w23 + w33),
                          w31 / (w11 + w21 + w31), w32 / (w12 + w22 + w32), w33 / (w13 + w23 + w33)]).reshape(3,3)
e_hidden = np.dot(normalization, error)
e_hidden


# 은닉계층 오차 계산 - 단순계산
print()
print("**은닉계층 오차 - 단순계산**")
e_hidden1 = e1 * w11 / (w11+w21+w31) + e2 * w12 / (w12+w22+w32) + e3 * w13 / (w13+w23+w33)
e_hidden2 = e1 * w21 / (w11+w21+w31) + e2 * w22 / (w12+w22+w32) + e3 * w23 / (w13+w23+w33)
e_hidden3 = e1 * w31 / (w11+w21+w31) + e2 * w32 / (w12+w22+w32) + e3 * w33 / (w13+w23+w33)

e_hidden1
e_hidden2
e_hidden3
```

```python
# 에러값
e1 = e_hidden1
e2 = e_hidden2
e3 = e_hidden3

# 가중치
w11 = w_input_hidden[0][0]
w12 = w_input_hidden[1][0]
w13 = w_input_hidden[2][0]
w21 = w_input_hidden[0][1]
w22 = w_input_hidden[1][1]
w23 = w_input_hidden[2][1]
w31 = w_input_hidden[0][2]
w32 = w_input_hidden[1][2]
w33 = w_input_hidden[2][2]

print("**입력계층 오차 - 행렬곱**")
error = np.array([e1, e2, e3]).reshape(3,1)
normalization = np.array([w11 / (w11 + w21 + w31), w12 / (w12 + w22 + w32), w13 / (w13 + w23 + w33),
                          w21 / (w11 + w21 + w31), w22/ (w12 + w22 + w32), w23 / (w13 + w23 + w33),
                          w31 / (w11 + w21 + w31), w32 / (w12 + w22 + w32), w33 / (w13 + w23 + w33)]).reshape(3,3)
e_input = np.dot(normalization, error)
e_input


print()
print("**입력계층 오차 - 단순계산**")
# 입력게층 오차 계산 - 단순계산
e_input1 = e1 * w11 / (w11+w21+w31) + e2 * w12 / (w12+w22+w32) + e3 * w13 / (w13+w23+w33)
e_input2 = e1 * w21 / (w11+w21+w31) + e2 * w22 / (w12+w22+w32) + e3 * w23 / (w13+w23+w33)
e_input3 = e1 * w31 / (w11+w21+w31) + e2 * w32 / (w12+w22+w32) + e3 * w33 / (w13+w23+w33)

e_input1
e_input2
e_input3
```

```python
# 은긱계층 업데이트값 구하기

# 출력부 오차값
# e = e_output
e = np.array([[-0.71630335],[-0.69859807],[ 0.21190294]])

# 은닉->출력 가중치
w = np.array([.3, .7, .5, .6, .5, .2, .8, .1, .9]).reshape(3,3)

# 은닉->출력 입력값(1번 문제 참고)
o = np.array([[0.76133271],[0.60348325],[0.65021855]])

# 시그모이드 대입값
s = np.dot(w, o)

value = -e * sigmoid(s) * (1-sigmoid(s))
value

print("업데이트량 계산")
w_update = np.dot(value, o.T)
w_update

print()
print("은닉계층 업데이트 결과")
w - w_update
```
