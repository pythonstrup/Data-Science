```python
# 패키지 import
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import matplotlib.pyplot as plt
```

```python
x_tensor = torch.tensor(data=[2.0, 3.0], requires_grad=True)
print(x_tensor)
```

```python
# 경사하강법 간단한 예제
x = torch.tensor(data=[2.0, 3.0], requires_grad=True)
y = x**2
z = 2*y + 3

target = torch.tensor([3.0, 4.0])
loss = torch.sum(torch.abs(z - target))
loss.backward()
print(x.grad, y.grad, z.grad)
```

```python
num_data = 1000
num_epoch = 500
x = init.uniform_(torch.Tensor(num_data, 1), -10, 10)
noise = init.normal_(torch.FloatTensor(num_data, 1), mean=0.0, std=1.0)
y = 2*x + 3
y_noise = 2*(x + noise) + 3

model = nn.Linear(1, 1)
loss_func = nn.L1Loss()

optimizer = optim.SGD(model.parameters(), lr=0.01)

# 손실을 기록할 배열
loss_arr = []

# 목표값은 y_noise
label = y_noise

for i in range(num_epoch):
  # 이전 학습의 기울기를 지우고 최적화 함수를 초기화
  # 기울기를 지우지 않으면 기존의 업데이트 때문에 학습이 잘 되지 않음
  optimizer.zero_grad()

  # 입력값 x를 모델에 대입
  output = model(x)

  # 결과값과 목표값의 차이를 L1 손실함수로 구하기
  loss = loss_func(output, label)

  # 손실에 대한 기울기
  loss.backward()

  # 구한 기울기를 이용해 모델의 변수를 업데이트
  optimizer.step()

  # 10번마다 모델의 변수가 어떻게 변하는지 출력
  if i % 10 == 0:
    # 
    plt.scatter(x.detach().numpy(), output.detach().numpy())
    plt.axis([-10, 10, -30, 30])
    plt.grid()
    plt.show()
    print(loss.data)

  loss_arr.append(loss.detach().numpy())
```

```python
plt.figure(figsize=(12,9))
plt.scatter(x.detach().numpy(), y_noise.detach().numpy(), color="red", edgecolor="black")
plt.scatter(x.detach().numpy(), output.detach().numpy(), color="black")
plt.axis([-10, 10, -30, 30])
plt.grid()
plt.show()

plt.figure(figsize=(15,15))
plt.scatter(x.numpy(),y_noise.numpy(),s=5,c="gray")
plt.scatter(x.detach().numpy(),output.detach().numpy(),s=5,c="red")
plt.axis([-10, 10, -30, 30])
plt.grid()
plt.show()
```

```python
import numpy as np

x = np.array([x for x in range(len(loss_arr))])
y = loss_arr

plt.plot(x, y)
plt.grid()
plt.show()
```
