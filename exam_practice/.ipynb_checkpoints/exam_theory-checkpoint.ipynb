#### 1. 보기의 딥러닝 모델들의 개념도와 모델을 짝지으시오. [각 1 점]
#### 2. 보기의 딥러닝 모델들에 대하여 전혀 모르는 사람도 이해할 수 있도록 자신만의 표현으로 설명하시오. [각 1 점]

<br/>

1. ANN (Artificial neural network)
    - 입력에 가중치를 곱한 합을 활성화 함수에 통과시키는 간단한 형태
    - 나만의 정리: 사람의 신경망 원리와 구조를 모방하여 만든 기계학습 알고리즘이다. 뉴런이 특정 자극이 임계값(threshold)을 넘어야지만 결과 신호를 전달하는 것처럼 인공신경망에서 가중치를 임계값에 빗대 만든 뉴럴 네트워크

<br/>

2. DNN (Deep neural network)
    - 은닉 계층이 2개 이상
    - 나만의 정리: ANN의 문제를 해결하기 위해 기존 neural network 모델에 은닉층을 2개 이상으로 늘려 학습의 결과를 향상시키는 심층 신경망이다.

<br/>

3. CNN (Convolutional neural network)
    - 컨볼루션 은닉 계층을 포함
    - 나만의 정리: 합성곱을 이용한 모델로, 데이터를 1차원으로 만들어 인식하는 완전 연결 네트워크와는 다르게 이미지의 공간정보를 유지한채 학습을 진행하는 네트워크이다. image 전체를 보는 것이 아니라 부분을 보는 필터를 이용해 특성지도(feature map)을 만들어 인식한다.

<br/>

4. RNN(recurrent neural network)
    - 순환적으로 들어오는 입력에 따라 새로운 은닉 벡터를 순환 생성
    - 나만의 정리: 반복적이고 순차적인 데이터 학습에 유리한 인공신경망이다. 내부의 순환구조를 가지고 있어 과거의 학습을 현재 학습에 반영할 수 있는 뉴럴 네트워크다.
    
<br/>
    
5. AE (AutoEncoder)
    - 특징의 압축과 복원
    - 나만의 정리: 입력을 출력으로 복사하는 신경망으로 hidden layer의 크기를 input layer보다 작게 만들어 데이터를 압축하거나, 입력값에 노이즈를 추가해 원본 입력을 복원할 수 있도록 학습시키는 네트워크이다.

<br/>

6. GAN (Generative adversarial network)
    - 생성망과 판별망이 서로 경쟁하며 원본을 복제
    - 나만의 정리: GAN은 서로 대립하는 2개의 시스템이 서로 경쟁하면서 가짜 이미지를 진짜 이미지와 최대한 비슷하게 만들어내는 네트워크이다. A가 가짜 이미지를 만들고 B가 그것을 감별한다면 B가 가짜 이미지를 구분할 수 없을 때까지 학습시킨다.

<br/>

7. DQN (Deep Q-Network)
    - 현 상태에서의 행동이 환경과 상호작용하여 보상 받고 새로운 상태 생성 
    - 나만의 정리: 강화학습과 관련된 네트워크로  experience replay, target network를 적용해 효율성을 높인 것이다. 큐러닝과 딥러닝을 합친 것이며 큐테이블(Q-table) 대신 신경망(neural net)을 사용해서, 뉴럴넷이 Q 가치를 근사해낼 수 있도록 학습시키는 것이다. DQN은 근사 함수(approximating functioin)라고 볼 수 있다.


<br/>

3. CNN 모델에서는 다양한 컨볼루션 필터를 사용할 수 있다. 아래 각 항목과 같은 컨볼루션 필터(커널)를 아래와 다음 이미지에 적용했을 때 얻어지는 특성 맵(feature map)의 한변의 크기는? [각 1 점]



- 입력 이미지의 크기 = I
- 필터의 크기 = K
- 스트라이드 = S
- 특성지도의 크기 $O = floor(\frac{I - K}{S} + 1)$
- 패딩 = P
- 패딩을 넣은 입력의 특성지도 크기 $O = floor(\frac{I - K + 2P}{S} + 1)$
- floor는 바닥함수를 의미

① 1×1 필터, stride=1 <br/>
=> $O = floor(\frac{7 - 1}{1} + 1) = 7$ <br/>
=> 7x7에서 한 변의 크기는 7 <br/>
<br/>

② 1×1 필터, stride=2 <br/>
=> $O = floor(\frac{7 - 1}{2} + 1)$ = 4 <br/>
=> 4x4에서 한 변의 크기는 4 <br/>
<br/>

③ 3×3 필터, stride=1 <br/>
=> $O = floor(\frac{7 - 3}{1} + 1) = 5$ <br/>
=> 5x5에서 한 변의 크기는 5 <br/>
<br/>

④ 3×3 필터, stride=2  <br/>
=> $O = floor(\frac{7 - 3}{2} + 1) = 3$ <br/>
=> 3x3에서 한 변의 크기는 3 <br/>
<br/>

⑤ 5×5 필터, stride=1 <br/>
=> $O = floor(\frac{7 - 5}{1} + 1) = 3$ <br/>
=> 3x3에서 한 변의 크기는 3 <br/>
<br/>
 
⑥ 5×5 필터, stride=2 <br/>
=> $O = floor(\frac{7 - 5}{2} + 1) = 2$ <br/>
=> 2x2에서 한 변의 크기는 2 <br/>
<br/>

⑦ 7×7 필터, stride=1 <br/>
=> $O = floor(\frac{7 - 7}{1} + 1) = 1$ <br/>
=> 1x1에서 한 변의 크기는 1 <br/>
